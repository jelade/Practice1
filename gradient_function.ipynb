```python
import plotly.graph_objects as go
import numpy as np

# Define the function to plot
def f(x, y):
    # write the function you want to plot.
    return x*y +1

def interactive_2D_plot(f):
    # Generate the x, y coordinates to evaluate the function
    x = np.linspace(-5, 5, 50)
    y = np.linspace(-5, 5, 50)
    X, Y = np.meshgrid(x, y)

    # Evaluate the function at each point in the x, y grid
    Z = f(X, Y)

    # Create the plotly figure object and add the surface trace
    fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z)])

    # Set the axis labels
    fig.update_layout(scene=dict(xaxis_title='x', yaxis_title='y', zaxis_title='f(x,y)'))

    # Show the plot
    fig.show()

interactive_2D_plot(f)
```

```python
import matplotlib.pyplot as plt
xlist = np.linspace(-2.0, 2.0, 100)
ylist = np.linspace(-2.0, 2.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = f(X,Y)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X,Y,Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.add_patch(plt.Circle(( 0 , 0 ), 1, color = "black" , fill =False))
ax.add_patch(plt.Circle(( 0 , 0 ), 1.43, color = "black" , fill =False))
ax.set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```
